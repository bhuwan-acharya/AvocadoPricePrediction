```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import random
from fbprophet import Prophet
```

```python
avocardo_df = pd.read_csv('avocado.csv')
```

```python
avocardo_df.head()
```

```python
avocardo_df = avocardo_df.sort_values('Date')
```

```python
avocardo_df.head()
```

```python
plt.figure(figsize = (10,10))
plt.plot(avocardo_df['Date'], avocardo_df['AveragePrice'])
```

```python
plt.figure(figsize = [25,12])
sns.countplot(x = 'region',data=avocardo_df)
plt.xticks(rotation = 45)
```

```python
sns.countplot(x='year', data= avocardo_df)
```

```python
avocardo_prophet_df = avocardo_df[['Date','AveragePrice']]
```

```python
avocardo_prophet_df
```

```python
avocardo_prophet_df = avocardo_prophet_df.rename(columns ={'Date':'ds','AveragePrice':'y'})
```

```python
model = Prophet()
model.fit(avocardo_prophet_df)
```

```python
#forecast the future
future = model.make_future_dataframe(periods = 365)
forecast = model.predict(future)
```

```python
forecast
```

```python
figure = model.plot(forecast, xlabel = 'Date', ylabel = 'Price')
```

```python
figure = model.plot_components(forecast)
```

# Making the prediction for the specific region only

```python
avocardo_df
```

```python
avocardo_df_west = avocardo_df[avocardo_df['region']=='West']
```

```python
avocardo_df_west
```

```python
plt.figure(figsize = [20,20])
plt.plot(avocardo_df_west['Date'],avocardo_df_west['AveragePrice'])
```

```python
avocardo_df_west=avocardo_df_west.rename(columns = {'Date':'ds','AveragePrice':'y'})
```

```python
avocardo_df_west = avocardo_df_west[['ds','y']]
```

```python
m_west = Prophet()
m_west.fit(avocardo_df_west)
```

```python
future = m_west.make_future_dataframe(periods = 500)
forecast = m_west.predict(future)
```

```python
figure = m_west.plot(forecast, xlabel = 'Date', ylabel = 'Price')
```

```python
figure = m_west.plot_components(forecast)
```
